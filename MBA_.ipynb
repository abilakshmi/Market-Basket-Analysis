**APRIORI** **ALGORITHM**

**For using aprori need to convert data in list format..**

```python
transactions = []
for i in range(0,len(df)):
    transactions.append([str(df.values[i,j]) for j in range(0,20) if str(df.values[i,j])!='0'])
transactions[12]
```

**Call apriori function which requires minimum support, confidance and lift**

```python
rules = apriori(transactions,min_support=0.003,min_confidance=0.2,min_lift=3,min_length=2)
```

**To generates a set of rules in a generator file.**

```python
rules
```

**All rules need to be converted in a list.**

```python
Results = list(rules)
Results
```

```python
for item in Results:
    pair = item[0]
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])

    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```

**Convert result in a dataframe for further operation.**

```python
df_results = pd.DataFrame(Results)
df_results.head()
```

**Keep support in a separate data frame so we can use later**

```python
support = df_results.support
```

**All** **four** **empty** **list** **which** **will** **contain** **lhs**, **rhs**, **confidance** **and** **lift**

```python
first_values = []
second_values = []
third_values = []
fourth_value = []
#loop number of rows time and append 1 by 1 value in a separate list..
#first and second element was frozenset which need to be converted in list
for i in range(df_results.shape[0]):
    single_list = df_results['ordered_statistics'][i][0]
    first_values.append(list(single_list[0]))
    second_values.append(list(single_list[1]))
    third_values.append(single_list[2])
    fourth_value.append(single_list[3])
```

**convert all four list into dataframe**

```python
lhs = pd.DataFrame(first_values)
rhs= pd.DataFrame(second_values)
confidance=pd.DataFrame(third_values,columns=['Confidance'])
lift=pd.DataFrame(fourth_value,columns=['lift'])
```

**All four list together in a single dataframe**


```python
df_final = pd.concat([lhs,rhs,support,confidance,lift], axis=1)
df_final
```

**Set column name**

```python
df_final.columns = ['lhs',1,0,1,2,'rhs','support','confidance','lift']
```

**Drop 1,0,1 columns for further operation**

```python
df_final.drop(columns=[1,0,1,2],inplace=True)
```

**USING TRANSACTION ENCODER**

**Reducing the shape of the dataset**

```python
dataset = [['citrus fruit',	'semi-finished bread','margarine	ready soups'],['tropical fruit'	,'yogurt','coffee'],['pip fruit','yogurt','cream cheese',	'meat spreads'],['other vegetables','whole milk','condensed milk','long life bakery product'],['whole milk','butter','yogurt','rice'],['other vegetables','UHT-milk','rolls/buns','bottled beer'],['whole milk','cereals'],['butter milk','pastry'],['bottled water','canned beer']]
```

```python
print(dataset)
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_array = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_array, columns=te.columns_)
df.shape
```

```python
print(df)
```

```python
df.shape
```

```python
# Compute the support
support = df.mean()
support = pd.DataFrame(support, columns=['support']).sort_values('support',ascending=False)

# Print the support
support.head()
```

```python
support.describe()
```

**The items and itemsets with at least 5% support:**

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets_ap = apriori(df, min_support=0.01, use_colnames=True)
```

```python
print(frequent_itemsets_ap)
```

**SELECTING AND FILTERING ITEMS**

**Flitering the items based on support**

```python
frequent_itemsets = apriori(df, min_support = 0.05, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

**Flitering the items based on length and support more than 10%**

```python
frequent_itemsets[ (frequent_itemsets['length'] == 2) &
                   (frequent_itemsets['support'] >= 0.01) ]
```

```python
frequent_itemsets[ (frequent_itemsets['length'] == 1) &
                   (frequent_itemsets['support'] >= 0.01) ]
```

**ASSOSIATION RULE MINING**

```python
from mlxtend.frequent_patterns import association_rules
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

```python
rules.describe()
```

**Filter the dataframe  for a large lift (4) and high confidence (0.5)**

```python
rules[ (rules['lift'] >= 4) &
      (rules['confidence'] >= 1)&(rules['support'] >= 0.1) ]
```
